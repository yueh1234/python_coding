```python
import cv2 as cv
import numpy as np

img = cv.imread('data/test.JPG')
```

## accessing and modifying pixel values

```python

px = img[100,100,2]
print(px)
img.item(100,100,2)
img.itemset((100,100,2),100)
img.item(100,100,2)
```

## Accessing image properties

```python
print(img.shape) # rows, columns, channels
print(img.size) # total pixels
print(img.dtype) # data type, important due to easy buggin !!!
```

```python
# get point
def get_point(event,x,y,_,__):
    if event == cv.EVENT_LBUTTONDOWN:
        print(y,x)
cv.setMouseCallback('img',get_point)

while True:
    cv.imshow('img',img)
    if cv.waitKey(0) & 0xFF == 27:
        break
```

```python
# replace regions
eye = img[1350:1650,1570:1870]
img[0:300,0:300] = eye
img[:,:,2] = 0
while True:
    cv.imshow('eye',eye)
    cv.imshow('img',img)
    if cv.waitKey(0) & 0xFF == 27:
        break
```

```python
# splliting and mergin image channels
b,g,r = cv.split(img)
img = cv.merge(b,g,r)
```

# create border

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

BLUE = [255,0,0]

img1 = cv.imread('data/123.jpg')
replicate = cv.copyMakeBorder(img1,100,100,100,100,cv.BORDER_REPLICATE)


# replicate = cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_REPLICATE)
reflect = cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_REFLECT)
reflect101 = cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_REFLECT_101)
wrap = cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_WRAP)
constant= cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_CONSTANT,value=BLUE)
plt.subplot(231),plt.imshow(img1),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```
